### Annotation:

```python
text_combined = str(tweets.tweet_full_text)
```

```python
doc = nlp(text_combined)
```

### Tokenization 

```python
for token in doc:
    print(token)
```

```python
for token in doc:
    print(token.text, token.pos_) #pos_ : Parts of speech
```

```python
nouns = list(doc.noun_chunks)
nouns
```

```python
list(doc.sents)
```

```python
for ent in doc.ents:
    print(ent.text,ent.label_)
```
